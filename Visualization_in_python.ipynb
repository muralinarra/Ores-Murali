```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dengue = pd.read_csv("dengue2.csv")
dengue
```

```python
dengue.columns
```

```python
dengue.isna().sum()
```

```python
dengue2= dengue.fillna(0)
dengue2
dengue2.isna().sum()
```

```python
dengue2
```

```python
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x="dengue.current_temp", y="dengue.platelet", data=dengue2)
sns.regplot(x="dengue.current_temp", y="dengue.platelet", data=dengue2, scatter=False, lowess=True, line_kws={'color': 'red'})
plt.title('Scatter Plot with Loess Regression Line and Shaded Confidence Bands')
plt.xlabel('Current Temperature')
plt.ylabel('Platelet Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(dengue2['dengue.platelet'], kde=True, color='skyblue')
plt.title('Platelet Count Histogram')
plt.xlabel('Platelet Count')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="dengue.dengue", y="dengue.hemoglobin", data=dengue2)
plt.title('Box Plot of Hemoglobin Levels')
plt.xlabel('Dengue Status')
plt.ylabel('Hemoglobin Level')
plt.show()
```

```python
dengue2['dengue.dengue'] = dengue2['dengue.dengue'].astype(str)

plt.figure(figsize=(10, 6))
sns.lineplot(x='dengue.dengue', y='dengue.current_temp', data=dengue2)
plt.title('Line Chart of Current Temperature by Dengue Status')
plt.xlabel('Dengue Status')
plt.ylabel('Current Temperature')
plt.show()
```

```python
dengue2.to_csv("denguelatest.csv", index = False)
```
