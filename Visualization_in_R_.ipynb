```r
# Load necessary libraries
library(ggplot2)
library(dplyr)
```

```r
dengue <- read.csv("denguelatest.csv")
dengue

ggplot(dengue, aes(x=dengue.residence, y=dengue.hemoglobin, group=1, color=dengue.residence)) +
  geom_line() +
  labs(title='Line Chart of Hemoglobin Levels by Residence',
       x='Residence',
       y='Hemoglobin Level') +
  theme_minimal() +
  theme(legend.position="top")
```

```r

ggplot(dengue, aes(x=dengue.joint_muscle_aches, y=dengue.nausea_vomiting, color=dengue.dengue)) +
  geom_point() +
  labs(title='Scatter Plot of Joint Muscle Aches vs Nausea Vomiting',
       x='Joint Muscle Aches',
       y='Nausea Vomiting') +
  theme_minimal() +
  theme(legend.position="top")
```
