```python
import pandas as pd
import numpy as np

# Create a sample DataFrame (replace this with your actual dataset)
data = pd.DataFrame({
    'value': [10, 12, 11, 15, 200, 8, 9, 14, 13, 250]
})

# Define a function to handle outliers using Winsorization
def handle_outliers_winsorize(data, column, lower_percentile, upper_percentile):
    lower_bound = np.percentile(data[column], lower_percentile)
    upper_bound = np.percentile(data[column], upper_percentile)
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
    return data

# Set the percentiles for winsorization (e.g., 1st and 99th percentiles)
lower_percentile = 1
upper_percentile = 99

# Apply winsorization to handle outliers
data = handle_outliers_winsorize(data, 'value', lower_percentile, upper_percentile)

# Now, 'data' contains outliers handled using winsorization
print(data)
```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame (replace this with your actual dataset)
data = pd.DataFrame({
    'value': [10, 12, 11, 15, 20, 8, 9, 14, 13, 25],
    'category': ['A', 'B', 'A', 'B', 'C', 'C', 'A', 'B', 'C', 'A']
})

# Log transformation
data['log_value'] = np.log(data['value'])

# Standardization (Z-score scaling)
data['z_score_value'] = (data['value'] - data['value'].mean()) / data['value'].std()

# Min-Max scaling
min_value = data['value'].min()
max_value = data['value'].max()
data['min_max_value'] = (data['value'] - min_value) / (max_value - min_value)

# One-hot encoding
data = pd.get_dummies(data, columns=['category'], prefix=['category'])


# Binning (dividing 'value' into three bins)
data['value_bin'] = pd.cut(data['value'], bins=3, labels=['Low', 'Medium', 'High'])

# Display the transformed data
print(data)
```
